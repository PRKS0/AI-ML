```python
# question: # Calculate number of payes to talako pasal on time range: morning, afternoon, evening, night
```

```python
import pandas as pd
import numpy as np
```

```python
# loading and cleaning
df = pd.read_excel("D:/Esewa Statement Report.xlsx", header=8, skipfooter=8, parse_dates=['Date Time'])
df
```

```python
# get all the payment to talako pasal
key_word = ['family store', 'ninglasaini', 'ramesh bohara', 'ninglashaini']
pattern = '|'.join(key_word)

Paid = df[df['Dr.']>0]
talako_pasal = Paid[Paid['Description'].str.contains(pattern, case=False)]
talako_pasal
```

```python
# date time and amount
talako_pasal[['Date Time', 'Dr.']]
```

```python
# playing with time
# morning: 4am-11am (4-11)
# afternoon: 12pm-4pm (12-16)
# evening: 5pm-8pm (17-20)
# night: 9pm-3am (21-3)

hours = talako_pasal['Date Time'].dt.hour

time_calc = pd.DataFrame(data=hours)
time_calc['Amount'] = talako_pasal['Dr.']
time_calc.rename(columns={'Date Time': 'Hours'}, inplace=True)


bins = [3, 11, 16, 20, 21]
labels = ['4-11', '12-16', '17-20', '21-3']


time_calc['Hour Group'] = pd.cut(time_calc['Hours'], bins=bins, labels=labels, right=True)
group = time_calc.groupby('Hour Group')['Hours']
group.count()
```
