make new data frame/table for pasal except talako pasal. Table caontains Description/Name, Count/number of payes, Amount

keyword: [talako pasal, store, khaja ghar, restaurant, bhojanalaya, cottage, traders, hotel]

```python
import pandas as pd
import numpy as np

pd.set_option('display.width', 200)
```

```python
df = pd.read_excel("D:/Esewa Statement Report.xlsx", skiprows=8, skipfooter=8, parse_dates=['Date Time'])
df
```

```python
# separating talako pasal

keyword = ['family store', 'ninglasaini', 'ramesh bohara', 'ninglashaini']

    
pattern = '|'.join(keyword)

Dr_Data = df[df['Dr.']>0]

mask = Dr_Data['Description'].str.contains(pattern, case=False)
filtered = Dr_Data[mask]

print(filtered)


# grouping
result = (filtered.groupby('Description').agg({
    'Description': 'count',
    'Dr.': 'sum'
}).rename(columns={'Description': 'Name', 'Dr.': 'Amount', 'Description': 'Count'}))


print(result.shape)
result

```

```python
# removing talako pasal
other_keyword = ['store', 'khaja ghar', 'restaurant', 'bhojanalaya', 'cottage', 'traders', 'hotel']
pattern2 = '|'.join(other_keyword)
new_df = df[~df['Description'].str.contains(pattern, case=False)]
print(new_df)

mask2 = new_df['Description'].str.contains(pattern2, case=False)
filtered2 = new_df[mask2]

result2 = filtered2.groupby('Description').agg({
    'Description': 'count',
    'Dr.': 'sum'
}).rename(columns={'Dr.': 'Amount', 'Description':'Count'})

result2
```
