```python
df = pd.read_excel("D:/Esewa Statement Report.xlsx", skiprows=8, skipfooter=8, parse_dates=['Date Time'])
DrData = df[df['Dr.']>0]
CrData = df[df['Cr.']>0]
CrData
```

```python
# debit credit summary
Debit = df['Dr.'].sum()
print(f'Total Debit: {Debit}')

Credit = df['Cr.'].sum()
print(f'Total Debit: {Credit}')

# Highest and Lowest paid in
Hdebit = DrData['Dr.'].max()
Ldebit = DrData['Dr.'].min()
# Ldebit = df.loc[df['Dr.']>0, 'Dr.']
# Ldebit = df.loc[df['Dr.'] != 0, 'Dr.'].min()
print(f'Highest Paid Out: {Hdebit}')
print(f'Lowest Paid Out: {Ldebit}')

# Highest and Lowest Paid in
Hcredit = CrData['Cr.'].max()
Lcredit = CrData['Cr.'].min()
print(f'Highest Paid In: {Hcredit}')
print(f'Lowest Paid In: {Lcredit}')


```

```python
# different deposit
paid = DrData.groupby('Description')['Dr.'].count().sort_values(ascending=False)
paid.sort_values(ascending=False).sort_values(ascending=False)
paid
```

```python
# make new data frame/table for talako_pasal. Table contains Description/Name, Count/number of payes, Amount
unique_desc = DrData['Description'].unique()                        # array
count_unique_desc = DrData['Description'].value_counts()            # series
amount_unique_desc = DrData.groupby('Description')['Dr.'].sum()     # series
key_word = ['family store', 'ninglasaini', 'ramesh bohara', 'ninglashaini']

talako_pasal = pd.DataFrame(columns=['Description', 'Count', 'Amount'])

for name in unique_desc:
    for word in key_word:
        if word in name.lower():
            talako_pasal.loc[len(talako_pasal)] = [name, count_unique_desc[name], amount_unique_desc[name] ]

print(talako_pasal)
```
