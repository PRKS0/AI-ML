combine talako pasal and other pasal. only one row for talako pasal. Description: talako pasal, Count: sum of counts, Amount: sum of amounts

```python
import pandas as pd
import numpy as np
pd.set_option('display.width', 150)
pd.reset_option('display.max_rows')
```

```python
Main_Data = pd.read_excel("D:/Esewa Statement Report.xlsx", skiprows=8, skipfooter=8, parse_dates=['Date Time'])
df = Main_Data[Main_Data['Dr.']>0]
df
```

```python
pattern1 = '|'.join(['family store', 'ninglasaini', 'ramesh bohara', 'ninglashaini'])
mask1 = df['Description'].str.contains(pattern1, case=False)
talako_pasal = pd.DataFrame({
    'Name': ['talako_pasal'],
    'Count': [df.loc[mask1, 'Description'].count()],
    'Amount': [df.loc[mask1, 'Dr.'].sum()]
})


pattern2 = '|'.join(['store', 'khaja ghar', 'restaurant', 'bhojanalaya', 'cottage', 'traders', 'hotel', 'bhojanal'])
# new_df = df[~df['Description'].str.contains(pattern, case=False)]
without_talako_pasal = df[~df['Description'].str.contains(pattern1, case=False)]

mask2 = without_talako_pasal['Description'].str.contains(pattern2, case=False)
filtered2 = without_talako_pasal[mask2]

aru_pasal = filtered2.groupby('Description').agg(
    Count = ('Description', 'count'),
    Amount = ('Dr.', 'sum'),
).reset_index().rename(columns={'Description': 'Name'})

total_pasal = pd.concat([talako_pasal, aru_pasal], ignore_index=True)

total_pasal.sort_values(by='Count', ascending=False)
```
