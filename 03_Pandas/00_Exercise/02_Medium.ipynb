```python
import pandas as pd
pd.set_option('display.width', 150)
```

```python
# sort the data frame by "Salary" in descending order
df = pd.read_csv('../../../Data.csv', parse_dates=['Joining_Date'])

df.sort_values(by='Salary', ascending=False)
```

```python
# Calculate the average salary for each department
df.groupby('Department')['Salary'].mean()
```

```python
# Find the number of employees who work remotely
(df['Remote_Work'] == 'Yes').sum()
df['Remote_Work'].eq('Yes').sum()
```

```python
# Filter employee with performance scor as excellent and salary above 80000
df[(df['Salary']>80000) & (df['Performance_Score'] == 'Excellent')]
```

```python
# add new column as Tax 10% of salary
tax = df['Salary']*0.1
df['Tax'] = tax
df
```

```python
# group employees by gender and find their average age
df.groupby('Gender')['Age'].mean()
```

```python
# find the top 5 highest paying employee in finance department
top5 = df[df['Department']=='Finance'].nlargest(5, 'Salary')
top5
```

```python
# count the number of employees who joined after 2020
df[df['Joining_Date']>='2023-01-01']
```

```python
# replace poor performance scores with 'needs imporvement'
df.loc[df['Performance_Score'] == 'Poor', 'Performance_Score'] = 'Needs Improvement'
df
```

```python
# find the medain bonus for each department
df.groupby('Department')['Bonus'].median()
```
