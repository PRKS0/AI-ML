### Data Loading and Saving

```python
import numpy as np
import pandas as pd
```

```python
# SEP
customer_data = pd.read_csv('data/customers-100.csv', sep=',')
customer_data.head(10)

# by default pandas put sep value (,) for csv file. but when we are importing TSV (tab), Pipe separated value (|), we have to specify the sep value.
```

```python
# Header: Determine which row should be the header (column name)
customer_data = pd.read_csv('data/customers-100.csv', header=2)
customer_data.head(10)

# 0: first row of data (default)
# 1: now it treat second row as header
# None: give formal indexing  (0, 1, 2, ....)
```

```python
# index_col: determine which column should be the column-index of dataframe 
customer_data = pd.read_csv('data/customers-100.csv', index_col=None)
customer_data.head(5)

# None: 0, 1, 2, 3, 4 .. (default)
# 0: add first column as index
# 4: add fifth column as index 
```

```python
# usecols: to load specific columns
column2 = pd.read_csv('data/customers-100.csv', usecols=['First Name', 'City'])
column2
```

```python
# other
df = pd.read_csv('data/customers-100.csv', parse_dates=['Subscription Date'], na_values=['NA', '?'])
df['Subscription Date']

# parse-dates: convert subscription date column into type-date for further operations
# na_values: treat 'NA' and '?' values as NaN type. 
```

```python
# all in one 
olympics = pd.read_csv('data/olympics-data.csv', header=0, usecols=['name', 'born_date', 'born_city', 'NOC', 'weight_kg'], parse_dates=['born_date'])
olympics
```

```python
# writing data 
olympics.to_csv('data/olympics-data-modified.tsv', sep='\t', index=False)
```
