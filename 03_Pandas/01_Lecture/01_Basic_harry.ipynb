### Two types of data structure in pandas
* Series - 1D
* Dataframes - 2D

### Series: Just like list, but have labeled data. 

```python
import pandas as pd
```

```python
data1 = pd.Series([2, 4, 6, 8, 10])
print("Series:\n", data1)

# we can change those label/index. by defualt index is 0,1,2,3,4...
data2 = pd.Series([2, 3, 5, 7, 11], index=['a', 'b', 'c', 'd', 'e'])
print('\n',data2)

# combination of series make data frame
```

### DataFrames

```python
df = pd.DataFrame({"name": ["Prakash", "Darshan", "Nishant"], "marks": [90, 80, 70]})
print("Data frame using Python dictionary\n", df)

index = ["prime", "even", "odd"]
columns = ["num1", "num2", "num3", "num4", "num5"]
df = pd.DataFrame([[2, 3, 4, 5, 11], [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]], index= index, columns=columns)
print("\n Data frame using list \n", df)
```

### when we run function to our data frame, it return new data frame. original is as it is. to make change on original, give "inplace=True" parameter

```python
# what if we don't want those rows which have some null value. (gives new data frame)
df.dropna() 

# but original variable df is not changed. 
print(df)

# if we give "inplace=True" this parameter in function, the main data frame (variable) will be changed
```

```python
# what if we want to replace those NaN with some 0.
#df.fillna(0)  # original is not changed

df.fillna(0, inplace=True)  # make change in original
print(df)
```

```python
# What if we want to change the column name
df = pd.read_csv('data/iris.csv')

df.rename(columns={"sepal.length": "SL", "sepal.width": "SW"})  # but original is not changed
```

```python
df.info()
```

```python
# what if we want sepal.length as int instead of float
df['sepal.length'] = df['sepal.length'].astype(int)
df.info()
```

```python
# what if we want new column in our data frame
df['count'] = [i for i in range(len(df))]
df
```

```python
# what if we want a new column according to the existing column
def fx(sl):
    return sl * sl

df['sepal.length.square'] = df['sepal.length'].apply(fx)

# sepal.length is provided as parameter in fx function, and we are taking it as sl

df
```

```python
# what if we want our data frame in file (csv, excel, other)

df.to_csv("data/export.csv", index=False)

# this will create a new csv file named "export" inside data folder, containing df's data.
# if we don't want to include the index of data frame in file then we use "index=False".
```

### Data

```python
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'age': [25, 32, 37, 29, 41],
    'department': ['HR', 'IT', 'IT', 'Finance', 'HR'],
    'salary': [50000, 62000, 70000, 52000, 80000]
}

df = pd.DataFrame(data, index=[1, 2, 3, 4, 5])
df
```

### LOC and ILOC (accessing row and column)

```python
# accesing column (series 1D)
df['name']

# accessing column (data frame 2D)
df[['name']]

# accessing rows/columns using iloc (only position)
df.iloc[0]
df.iloc[0:3]   # row at index 3 not included
df.iloc[:, 0]  # first column (series)
df.iloc[:, 0]  # first column (data frame)

# accessing rows/columns using loc (label/inde_name)
df.loc[1]
df.loc[1:4]  # row at index 4 is also included
df.loc[:, 'name']   # first column
df.loc[:, ['name']]   # first column (data frame)
```

### Filtering Data

```python
df
```

```python
# salary greater than 55K
df[df['salary']>55000]
```

```python
# Sorting
df.sort_values(by='salary')
df.sort_values(by='age')
```

```python
df.groupby('department')['salary'].mean()
df.groupby('department')['salary'].agg(['mean', 'sum', 'count'])

df = pd.DataFrame({
    'department': ['HR', 'IT', 'IT', 'Finance', 'HR'],
    'salary': [50000, 62000, 70000, 52000, 80000],
    'bonus': [2000, 3000, 2500, 1500, 4000]
})

# Average salary and bonus per department
df.groupby('department').mean()

```

### Pivot Table

```python
import pandas as pd

data = {
    'department': ['HR', 'IT', 'HR', 'Finance', 'IT'],
    'gender': ['F', 'M', 'F', 'M', 'M'],
    'salary': [50000, 62000, 80000, 52000, 70000]
}

df = pd.DataFrame(data)
df
```

```python
pd.pivot_table(df, index='department', values='salary', aggfunc='mean')
```
