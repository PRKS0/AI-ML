```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print(" *** Zomato Sales Record *** ")
print("\nSales Data Shape:", sales_data.shape)
print("\n Data of first 3 restaurant", sales_data[0:3])

# Total sales per year
print("Total sales per year:", np.sum(sales_data, axis=0))
total_sales =  np.sum(sales_data[:, 1:], axis=0)
print("TOtal sales:",total_sales)
print("Grand total sales:", np.sum(total_sales))


# Min sales of each resturant
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Min sales of each restaurant:", min_sales)

# Max sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print("Max sales per year:", max_sales)

# Average sales per resturant
avg_sales = np.average(sales_data[:, 1:], axis=1)
print("average sales per restaurant:", avg_sales)

# Cumsum
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

# graph
plt.figure(figsize=(8, 6))
plt.plot(sales_data[:,1:])
plt.title("Total sales per year")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
