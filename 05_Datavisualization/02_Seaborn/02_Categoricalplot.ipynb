### Categorial Data Plots
Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot

Let's go through examples of each!

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('Tips')
df
# here sex, smoker, day, time, and size are categorical data
```

```python
# countplot: take data for only 1 axis (x or y)
# let's count the number of male and female
# sns.countplot(df['sex'])          # this will show data on y axis, so
sns.countplot(x = df['sex'])
```

```python
# how many male and female are smoking and not smoking
sns.countplot(x = df['sex'], hue=df['smoker'])
# hue: compare male and female on which basis
```

```python
# barplot: takes data for both axis (quantative on y, qualitative on x)
sns.barplot(x=df['day'], y=df['tip'])
# value on y-axis is average/mean of tip of each day
```

```python
# estimater (default: mean)
# total tips on each day

sns.barplot(x=df['day'], y=df['tip'], estimator=np.sum)
#estimator: compare every day on the basis of sum of tip/estimator
```


Boxplot


```python
sns.boxplot(x = 'tip', y='day', data = df) 
```

violinplot

```python
sns.violinplot(x = 'day', y = 'total_bill', data = df)
```

stripplot

```python
sns.stripplot(x = 'day', y = 'total_bill', data = df)
```

swarmplot

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = df)
sns.violinplot(x = 'day', y = 'total_bill', data = df)
```

Creating violin and swarm plot on same canvas

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = df)
```
