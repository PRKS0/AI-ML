Regression: on y-axis, there is prediction value, and x-axis hold data on which we do prediction. we get scatter plot and the best fit line between scatter is regression plot. 

we create scatter plot and whole process is called regression.

we use lmplot here.


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# lets take tips data
tips = sns.load_dataset('tips')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# same data for male and female separetely
sns.lmplot(x='total_bill', y='tip', data=tips, hue = 'sex')
```

```python
# let's change what shape represent male and female 
sns.lmplot(x='total_bill', y='tip', data=tips, hue = 'sex', markers=['o', 'v'])
```
