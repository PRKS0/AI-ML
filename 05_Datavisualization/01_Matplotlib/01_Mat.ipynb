```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Basic


```python
x = np.linspace(0, 5, 11)
y = x ** 2

plt.plot(x, y)
plt.title('graph')
plt.xlabel('number')
plt.ylabel('square')
```

### subplots

```python
# plt.Subplot(row, column, select-plot)
plt.subplot(2, 2, 1)
plt.plot(x*2, y)

plt.subplot(2, 2, 2)
plt.plot(x-1, y/2)

plt.subplot(2, 2, 3)
plt.plot(x-6, y**2)

plt.subplot(2, 2, 4)
plt.plot(x+40, y**3)
```

### Object oriented way

```python
fig = plt.figure()                              # creating canvas

axis1 = fig.add_axes([0.1, 0.1, 1, 1])          
axis1.plot(x,y)

axis2 = fig.add_axes([0.2, 0.4, 0.3, 0.3])      # axes1([gap-from-left, gap-from-bottom, horizontal-size, vertical-size])     
axis2.plot(x,y)
```

```python
new_fig = plt.figure(figsize=(9,5), dpi=100)            # size: 1200 x 800, dpi: increase quality (default: 100)
new_axis = new_fig.add_axes([0.1,0.1, 1, 1])
new_axis.plot(x,y)
```

### Types of plot

```python
# Scatter: dot, dot, dot
# other types: boxplot, hist, 
plt.scatter(x, y)
```

### Saving plots

```python
raw_plot = plt.plot(y,x)
plt.savefig('basicplot.png')
```
